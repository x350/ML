```python
import numpy as np
import pandas as pd
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
geo_comm = pd.read_excel('geo_comment.xlsx')
geo_comm.head()
```

```python
geo_comm.shape
```

```python
plt.figure(figsize=(12,9))
plt.scatter(geo_comm['x'], geo_comm['y'], c=geo_comm['comment_class'])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
X = geo_comm[['x', 'y']]
Y = geo_comm['comment_class']
# Y.head(3)
X_tr = np.array(X)
X_tr
```

### Разделить город на районы безотносительно голосов 

```python
model_KMeans = KMeans(n_clusters=2)
```

```python
y_pred = model_KMeans.fit_predict(X_tr)
```

```python
plt.figure(figsize=(12,9))
plt.scatter(X_tr[:,0], X_tr[:,1], c=y_pred)
plt.show()
```

```python
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_tr, )
    k_inertia.append(clf_kmeans.inertia_)
```

```python
plt.plot(ks, k_inertia)
```

 ###  Скопления голосов "за" и "против" отдельно


```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.5, min_samples=5, metric='euclidean')
```

```python
X_neg = geo_comm[geo_comm['comment_class'] == -1]
X_neg.shape
```

```python
X_poz = geo_comm[geo_comm['comment_class'] == 1]
X_poz.shape
```

```python
npX_poz = np.array(X_poz[['x', 'y']])
npX_neg = np.array(X_neg[['x', 'y']])
```

#### Голоса "За"

```python
k_poz = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(npX_poz )
    k_poz.append(clf_kmeans.inertia_)
    
plt.plot(ks, k_poz)
```

```python
model_KMeans = KMeans(n_clusters=4)
y_poz_pred = model_KMeans.fit_predict(npX_poz)
plt.figure(figsize=(12,9))
plt.scatter(npX_poz[:,0], npX_poz[:,1], c=y_poz_pred)
plt.show()
```

#### Голса "Против"

```python
k_neg = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(npX_neg)
    k_neg.append(clf_kmeans.inertia_)
    
plt.plot(ks, k_neg)
```

```python
model_KMeans = KMeans(n_clusters=3)
y_neg_pred = model_KMeans.fit_predict(npX_neg)
plt.figure(figsize=(12,9))
plt.scatter(npX_neg[:,0], npX_neg[:,1], c=y_neg_pred)
plt.show()
```

#### За и против с помощью DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.01, min_samples=5, metric='euclidean')
```

```python
npY_poz = model.fit_predict(npX_poz)
```

```python
plt.figure(figsize=(12,9))
plt.scatter(npX_poz[:,0], npX_poz[:,1], c=npY_poz)
plt.show()
```

```python
npY_neg = model.fit_predict(npX_neg)
```

```python
plt.figure(figsize=(12,9))
plt.scatter(npX_neg[:,0], npX_neg[:,1], c=npY_neg)
plt.show()
```

### наиболее активные в голосовании

```python
model = DBSCAN(eps=0.002, min_samples=7, metric='euclidean')
npY_poz = model.fit_predict(npX_poz)
plt.figure(figsize=(12,9))
plt.scatter(npX_poz[:,0], npX_poz[:,1], c=npY_poz)
plt.show()
```
